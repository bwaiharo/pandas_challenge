```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

```python
pc = len(purchase_data['SN'].unique())
player_count = pd.DataFrame({"Total Players": [pc] })
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# purchase_data.columns
# purchase_data.tail()
unique_items = len(purchase_data['Item Name'].unique())
avg_purchase_price = purchase_data['Price'].mean()
total_revenue = purchase_data['Price'].sum()
number_of_purchases = len(purchase_data['Purchase ID'])

pa =pd.DataFrame({"Number of Unique Items": [unique_items],
                  "Average Price": [avg_purchase_price],
                  "Number of Purchases":[number_of_purchases],
                  "Total Revenue": [total_revenue]})
# Proper formating
pd.options.display.float_format = '${:,.2f}'.format
pa

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Avg Total Purchase per Person
new_p =purchase_data.groupby(["Gender","SN"], as_index=False)['Price'].sum()
male_avg_total_per_person=new_p[new_p['Gender']=='Male']['Price'].mean()
female_avg_total_per_person=new_p[new_p['Gender']=='Female']['Price'].mean()
other_avg_total_per_person=new_p[new_p['Gender']=="Other / Non-Disclosed"]['Price'].mean()

# Average Purchase Price
male_avg_purchase = purchase_data[purchase_data["Gender"]=="Male"]["Price"].mean()
female_avg_purchase =purchase_data[purchase_data["Gender"]=="Female"]["Price"].mean()
other_avg_purchase =purchase_data[purchase_data["Gender"]=="Other / Non-Disclosed"]["Price"].mean()

# Total Purchase Value
male_total = purchase_data[purchase_data["Gender"]=="Male"]["Price"].sum()
female_total=purchase_data[purchase_data["Gender"]=="Female"]["Price"].sum()
other_total=purchase_data[purchase_data["Gender"]=="Other / Non-Disclosed"]["Price"].sum()

# Purchase Count
male_count = purchase_data[purchase_data["Gender"]=="Male"]["Price"].count()
female_count = purchase_data[purchase_data["Gender"]=="Female"]["Price"].count()
other_count = purchase_data[purchase_data["Gender"]=="Other / Non-Disclosed"]["Price"].count()

# Dataframe
pag = pd.DataFrame({"Gender":["Male","Female","Other / Non-Disclosed"],
                   "Purchase Count":[male_count,female_count,other_count],
                   "Average Purchase Price":[male_avg_purchase,female_avg_purchase ,other_avg_purchase],
                   "Total Purchase Value":[male_total,female_total,other_total],
                   "Avg Total Purchase per Person":[male_avg_total_per_person,female_avg_total_per_person,other_avg_total_per_person]})

# Proper formating
# pd.options.display.float_format = '${:,.2f}'.format

# Output
pag
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
ad = pd.DataFrame({     })

bins = [0,10,14,19,24,29,34,40,100]
age_group = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"] 
count_age= purchase_data.groupby(pd.cut(purchase_data['Age'],bins, right = False, labels = age_group))
ad["Total Count"] = (count_age['Age'].count())
ad["Percentage of Players"] = ((ad["Total Count"]/ad["Total Count"].sum())*100).map("{:.2f}%".format)

ad

```

* Create a summary data frame to hold the results
